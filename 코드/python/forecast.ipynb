```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

# 입력 데이터 생성
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # 1차식의 입력
y = np.array([3, 5, 7, 9, 11])

# 2차항으로 변환
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X = poly_features.fit_transform(x)

# 선형 회귀 모델 학습
model = LinearRegression()
model.fit(X, y)

# 새로운 데이터에 대한 예측
new_x = np.array([6]).reshape(-1, 1)
new_X = poly_features.transform(new_x)
predicted_y = model.predict(new_X)

# 예측 결과 그래프로 표시
plt.scatter(x, y, color='blue', label='Actual')  # 실제 데이터
plt.plot(new_x, predicted_y, color='red', label='Predicted')  # 예측 결과
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# 예측 결과 그래프로 표시 (새로운 데이터)
new_X = poly_features.transform(new_x)
predicted_y = model.predict(new_X)
plt.scatter(new_x, predicted_y, color='green', label='Predicted (New)')  # 새로운 데이터에 대한 예측 결과
plt.legend()
plt.show()
```
